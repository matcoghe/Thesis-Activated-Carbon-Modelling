```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(Ce, KF, n):
    return KF * Ce**(1/n)
```

```python
Cedata = np.linspace(0, 6, 50)
qe = func(Cedata, 2.7, 3)
rng = np.random.default_rng()
qe_noise = 0.2 * rng.normal(size=Cedata.size)
qedata = qe + qe_noise

plt.plot(Cedata, qedata, 'b-', label='data')
popt, pcov = curve_fit(func, Cedata, qedata)
popt
plt.plot(Cedata, func(Cedata, *popt), 'r-',
         label='fit: KF=%5.3f, n=%5.3f' % tuple(popt))
popt, pcov = curve_fit(func, Cedata, qedata, bounds=(0, [10., 10.]))

plt.xlabel('Ce')
plt.ylabel('qe')
plt.legend()
plt.show()
```
